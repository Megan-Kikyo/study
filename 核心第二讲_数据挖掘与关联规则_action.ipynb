```python
###针对MarketBasket数据集进行购物篮分析（频繁项集及关联规则挖掘）###
###采用mlxtend package###

#相关包导入#
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import os
import pandas as pd
```

```python
##数据读入与预处理##
os.chdir('D:/开课吧数据挖掘资料/第二讲—数据挖掘与关联规则/code/MarketBasket')
data = pd.read_csv('./Market_Basket_Optimisation.csv',header=None)
print(data.head())
print(data.shape)
```

```python
#存为id item dataframe， 同一个id下多个item 分为多行进行存取#    
temp=[]
temp = pd.DataFrame(columns = ['id','item'])    
for i in range(0,data.shape[0]):
    for j in range(0,data.shape[1]):
        if str(data.values[i,j])!='nan':
#            print(i,str(data.values[i,j]))
            temp1={}
            temp1['id']=i
            temp1['item']=str(data.values[i,j])
            temp=temp.append(temp1,ignore_index=True)
print(temp.head())
print(temp.shape) 
```

```python
# 哑变量编码 #            
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
# 采用mlxtend.frequent_patterns工具包  将数据集转变为 id item level上 保持唯一性；进行哑变量转换# 
hot_encoded_df=temp.groupby(['id','item'])['item'].count().unstack().reset_index().fillna(0).set_index('id') 
hot_encoded_df=hot_encoded_df.applymap(encode_units) 
```

```python
#频繁项集与关联规则挖掘#
frequent_itemsets = apriori(hot_encoded_df, min_support=0.03, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print("频繁项集：", frequent_itemsets)
print("关联规则：", rules[ (rules['lift'] >1) & (rules['confidence'] >= 0.2) ])
```
